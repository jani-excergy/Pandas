```python
import pandas as pd
```

```python
# Pandas is a python module that makes data science easy and effective
# It used for to read data and manupilation data 
# In pandas important data structure is Dataframe 
# Data frame is a data structure represent the data in tabular or excel spread sheet like data
```

```python
#dataframe = pd.DataFrame(data=iterable,index,columns)
#iterables : lists,tuples,dictonary

# Creating Dataframe by list ot tuples

data= [('Ram',98,94,95,92),('Robert',92,91,90,92),('Rahim',93,94,95,92)]
df=pd.DataFrame(data,columns=['Name','Maths','Physics','Chemistry','English'])
df
```

```python
# Creating Dataframe by Dictonary 

data= {'Maths':[98,92,93],'physics':[94,91,94],'chemistry':[95,90,95],'english':[92,92,92]}
df=pd.DataFrame(data,index=['Ram','Robert','Rahim'])
```

```python
# Creating Dataframe with numpy ndarray
import numpy as np

df = pd.DataFrame(np.array([[98,94,95,92], [92,92,90,92], [93,94,95,92]]),
                   columns=['Maths', 'Physics', 'chemistry','English'],index=['Ram','Robert','Rahim'])
df
```
