```python
# .apply() on each row of data frame

def main(): 
      
    
    data = { 
            'A':[1, 2, 3],  
            'B':[4, 5, 6],  
            'C':[7, 8, 9] } 
      
      
    df = pd.DataFrame(data) 
    print("Original DataFrame:\n", df) 
      
    
    df['add'] = df.apply(np.sum, axis = 1) 
   
    print('\nAfter Applying Function: ') 
   
    print(df) 
    
if __name__ == '__main__': 
    main()
```

```python
def normalize(x, y): 
    x_new = ((x - np.mean([x, y])) /
             (max(x, y) - min(x, y))) 
      
     
    return x_new 
  
def main(): 
      
     
    data = { 
        'X':[1, 2, 3],  
        'Y':[45, 65, 89] } 
      
      
    df = pd.DataFrame(data) 
    print("Original DataFrame:\n", df) 
      
    df['X'] = df.apply(lambda row : normalize(row['X'], 
                                  row['Y']), axis = 1) 
   
    print('\nNormalized:') 
    print(df) 
    
if __name__ == '__main__': 
    main() 
```
